```python
rdd = spark.sparkContext.textFile("README.md")
```

```python
words = rdd.flatMap(lambda line: line.split())
word_counts = words.map(lambda word: (word, 1)).reduceByKey(lambda a, b: a + b)
for word, count in word_counts.collect():
 print(f"{word}: {count}")
```

```python
output_file = "palavras_contadas.txt"

# Abrir o arquivo no modo de escrita
with open(output_file, "w") as f:
    for word, count in word_counts.collect():  # Certifique-se de que word_counts j√° est√° definido.
        f.write(f"{word}: {count}\n")  # Escrever cada palavra e sua contagem no arquivo.

# Fazer o download do arquivo no Google Colab
from google.colab import files
files.download(output_file)
```
